# CanESM5-r1i1p1f1 issues for ssp245 in the pangeo GCS catalogue

- As illustrated below, the data from the Omon table will not load properly when access from the pangeo GCS catalogue
- It has been confirmed that the original data published on ESGF works as expected, hence this appears to be a conversion issue.
- It has also been confirmed that other CanESM5 data from Pangeo GCS catalogue does work as expected, for example, other ssps, and even `r1i1p2f1` from ssp245

```python
cat_url = "https://storage.googleapis.com/cmip6/pangeo-cmip6.json"       # use this outside of CCCma / for public data
col = intake.open_esm_datastore(cat_url)
query = dict(experiment_id=['ssp245'], table_id=['Omon'], member_id='r1i1p1f1',
             variable_id=['fgco2'], source_id='CanESM5')
cat = col.search(**query)

ds_g = cat.to_dataset_dict(zarr_kwargs={"consolidated": True, "decode_times": True}, cdf_kwargs={'chunks': {'time':1032}})['ScenarioMIP.CCCma.CanESM5.ssp245.Omon.gn']
```
